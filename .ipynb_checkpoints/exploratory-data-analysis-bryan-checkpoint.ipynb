```python
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
sb.set_theme()
```

### 1. Import Cleaned dataset into the notebook

```python
# heart_pki_2020_cleaned_df
heart_pki_df = pd.read_csv('datasets/heart_pki_2020_cleaned.csv')
heart_pki_df.head()
```

```python
# heart_attack_cleanned_text_df
heart_attack_df = pd.read_csv('datasets/heart_attack_cleaned_text.csv')
heart_attack_df.head()
```

### 2. heart_pki_df EDA

```python
heart_pki_df.info()
```

```python
# helper function to show percentage when plotting categorical variables with cardio
def plotCatVar(x):
    order = heart_pki_df[x].sort_values().unique()
    hue_order = heart_pki_df['HeartDisease'].unique()
    ax= sb.countplot(x=x, hue='HeartDisease',data=heart_pki_df, order = order, hue_order = hue_order)

    plt.xticks(size=12)
    plt.yticks(size=12)
    ax.set_xticklabels(ax.get_xticklabels(), ha="right")

    total = len(heart_pki_df)
    for p in ax.patches:
        percentage = f'{100 * p.get_height() / total:.1f}%\n'
        x = p.get_x() + p.get_width() / 2
        y = p.get_height()+.05
        ax.annotate(percentage, (x, y), ha='center', va='center')
    plt.tight_layout()
    plt.show()
```

### Age

```python
fig,ax = plt.subplots(1,1,figsize=(16,12))
sb.countplot(x='AgeCategory',data=heart_pki_df)
```

```python
fig,ax = plt.subplots(1,1,figsize=(16,12))
plotCatVar('AgeCategory')
```

People of higher age have a higher chance of having Heart Disease
