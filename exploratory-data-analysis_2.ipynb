## 2.2 HeartAttack - Categorical Variables
- `Categorical`: sex, chest_pain, fasting_blood_sugar, rest_ecg, exercise_induced_angina, num_of_major_vessels, thal_rate

```python
# helper function to show percentage when plotting categorical variables with cardio
def plotCatVar(x, uniques_values):
    temp = heart_atk_clean_df[heart_atk_clean_df[x] != '0']
    ax = sb.countplot(x = temp[x], hue = temp['heart_attack_chance'], palette="coolwarm_r")

    total = len(temp[x])
    ind = 0
    for p in ax.patches:
        if ind<uniques_values == 0:
            total = ax.patches[ind].get_height() + ax.patches[ind+uniques_values].get_height()
        else:
            total = ax.patches[ind].get_height() + ax.patches[ind-uniques_values].get_height()
        percentage = f'{100 * ax.patches[ind].get_height() / total:.1f}%\n'
        x = p.get_x() + p.get_width() / 2
        y = p.get_height()
        ax.annotate(percentage, (x, y), ha='center', va='center')
        ind = ind+1
    plt.tight_layout()
    plt.show()
```

### 2.2.1 HeartAttack - Categorical Variables - `sex`

```python
male_df = heart_atk_clean_df[heart_atk_clean_df["sex"] == "Female"]
female_df = heart_atk_clean_df[heart_atk_clean_df["sex"] == "Male"]

male_hd = male_df['heart_attack_chance'].value_counts()
female_hd = female_df['heart_attack_chance'].value_counts()


fig, axs = plt.subplots(1,2,figsize=(12,5))
fig.add_subplot(axs[0])
plt.pie(male_hd, labels = male_hd.index,startangle = 90,counterclock = False, autopct = '%1.1f%%')
plt.title('Chances of Heart Attack (Male)')


fig.add_subplot(axs[1])
plt.pie(female_hd, labels = male_hd.index, startangle = 90,counterclock = False, autopct = '%1.1f%%')
plt.title('Chances of Heart Attack (Female)')
```

- Males have a higher chances of having a Heart Attack as compared to Females

### 2.2.2 HeartAttack - Categorical Variables - `chest_pain`

```python
plt.figure(figsize = (8,5))
plotCatVar('chest_pain',3)
```

- If someone has chest pain, the chances of them having a heart attack is over 69%
- The type of chest pain that results in a the highest chance of heart attack would be atypical anigna followed by typical angina and non-anginal  

### 2.2.3 HeartAttack - Categorical Variables - `fasting_blood_sugar`
- (fasting blood sugar > 120 mg/dl) (1 = true; 0 = false)

```python
no_fast_df = heart_atk_clean_df[heart_atk_clean_df["fasting_blood_sugar"] == False]
fast_df = heart_atk_clean_df[heart_atk_clean_df["fasting_blood_sugar"] == True]

no_fast = no_fast_df['heart_attack_chance'].value_counts()
fast = fast_df['heart_attack_chance'].value_counts()


fig, axs = plt.subplots(1,2,figsize=(12,5))
fig.add_subplot(axs[0])
plt.pie(no_fast, labels = no_fast.index,startangle = 90,counterclock = False, autopct = '%1.1f%%')
plt.title('People without fasting_blood_sugar')


fig.add_subplot(axs[1])
plt.pie(fast, labels = fast.index, startangle = 90,counterclock = False, autopct = '%1.1f%%')
plt.title('People with fasting_blood_sugar')
```

- Those without fasting blood sugar levels have a slightly higher chance of having a heart attack

### 2.2.5 HeartAttack - Categorical Variables - `exercise_induced_angina`
- exercise induced angina (1 = yes; 0 = no)

```python
less_chance_df = heart_atk_clean_df[heart_atk_clean_df["heart_attack_chance"] == "less chance"]
more_chance_df = heart_atk_clean_df[heart_atk_clean_df["heart_attack_chance"] == "more chance"]

less_chance = less_chance_df['exercise_induced_angina'].value_counts()
more_chance = more_chance_df['exercise_induced_angina'].value_counts()

colours = {'Yes': 'wheat',
           'No': 'rosybrown'}

fig, axs = plt.subplots(1,2,figsize=(12,5))
fig.add_subplot(axs[0])
plt.pie(less_chance, labels = less_chance.index,startangle = 90,counterclock = False, autopct = '%1.1f%%', colors=[colours[key] for key in less_chance.index])
plt.title('Percentage of people with less chance of heart attack\n that have exercise induced angina')


fig.add_subplot(axs[1])
plt.pie(more_chance, labels = more_chance.index, startangle = 90,counterclock = False, autopct = '%1.1f%%', colors=[colours[key] for key in more_chance.index])
plt.title('Percentage of people with more chance of heart attack\n that have exercise induced angina')
```

```python
sb.violinplot(y = heart_atk_df['heart_attack_chance'], x = heart_atk_df['exercise_induced_angina'])
```

- From the pie charts, we can see that those with less chance of havign a heart attack have a higher chance from suffering from exercise induced angina
- From the violin plot, it is shown that chances of having a heart attack is high when exercise_induced_angina is false and vice versa.

From the point of view of the clustering model:
- It shows the hierarchy of the variables after being clustered under the correlation perspective.
- Looking at the hierarchy from bottom to top: chest pain and maximal heart rate in one group, gender and chest rate in one group, age and number of MRI vessels in one group, and resting blood pressure and fasting glucose in one group.
- Variables in the same group mean: they are strongly correlated with each other, and variables in the same group are substitutable for each other.
