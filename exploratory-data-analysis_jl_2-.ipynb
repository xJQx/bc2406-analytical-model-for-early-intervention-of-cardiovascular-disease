<b> please format and edit language </b>
1. First of all, `age` and `max_heart_rate` demonstrate an obvious negative correlation; people tend to have a higher maximum heart rate when they are young, and the maximum heart rate will drop when they get older.
2. People who are prone to heart disease have a relatively high heart rate when they are young.
3. Those who are not prone to heart disease have little change in their maximum heart rate throughout their lives; however, those who are prone to heart disease have extremely high declines in maximum heart rate.

### 2.2.1 HeartAttack - Categorical Variables `sex`

```python
male_df = heart_atk_clean_df[heart_atk_clean_df["sex"] == "Female"]
female_df = heart_atk_clean_df[heart_atk_clean_df["sex"] == "Male"]

male_hd = male_df['heart_attack_chance'].value_counts()
female_hd = female_df['heart_attack_chance'].value_counts()


fig, axs = plt.subplots(1,2,figsize=(12,5))
fig.add_subplot(axs[0])
plt.pie(male_hd, labels = male_hd.index,startangle = 90,counterclock = False, autopct = '%1.1f%%')
plt.title('Chances of Heart Attack (Male)')


fig.add_subplot(axs[1])
plt.pie(female_hd, labels = male_hd.index, startangle = 90,counterclock = False, autopct = '%1.1f%%')
plt.title('Chances of Heart Attack (Female)')
```

- Discoveries

### 2.2.2 HeartAttack - Categorical Variables `chest_pain`

```python
temp = heart_atk_clean_df[heart_atk_clean_df['chest_pain'] != '0']
plt.figure(figsize = (8,5))
ax = sb.countplot(x = temp['chest_pain'], hue = temp['heart_attack_chance'], palette="coolwarm_r")

total = len(temp['chest_pain'])
ind = 0
for p in ax.patches:
    if ind<3 == 0:
        total = ax.patches[ind].get_height() + ax.patches[ind+3].get_height()
    else:
        total = ax.patches[ind].get_height() + ax.patches[ind-3].get_height()
    percentage = f'{100 * ax.patches[ind].get_height() / total:.1f}%\n'
    x = p.get_x() + p.get_width() / 2
    y = p.get_height()
    ax.annotate(percentage, (x, y), ha='center', va='center')
    ind = ind+1
plt.tight_layout()
plt.show()
```

Discoveris

From the point of view of clustering model.
- It shows the hierarchy of the variables after being clustered under the perspective of correlation.
- Looking at the hierarchy from bottom to top: chest pain and maximal heart rate in one group, gender and chest rate in one group, age and number of MRI vessels in one group, and resting blood pressure and fasting glucose in one group.
- Variables in the same group mean: they are strongly correlated with each other and variables in the same group are substitutable for each other.
